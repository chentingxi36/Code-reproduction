```python
# 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#  excelcsvmatlabpandasexcel
# pandascsvdf
df = pd.read_csv('41467_2018_6625_MOESM3_ESM.csv')
# 
plt.figure(figsize=(6, 6))
plt.scatter(df['Band Gap (eV)'], df['Debye T (K)'], edgecolor='black', linewidth=0.5, facecolors='gray', alpha=1, s=20)
plt.title('Band Gap vs Debye T')
plt.xlabel('Band Gap (eV)')
plt.ylabel('Debye T (K)')
plt.show()
```

```python
import re
# 
df['Contains_B_O'] = df['Formula'].apply(lambda x: 'B' in x and 'O' in x and 'Bi' not in x and 'Br' not in x and 'Ba' not in x)
df['Contains_S'] = df['Formula'].apply(lambda x: bool(re.search(r'S[^a-z]', x)))
#df['Contains_S'] = df['Formula'].apply(lambda x: 'S' in x and 'Sr' not in x and 'Sc' not in x and 'Sn' not in x and 'Sb' not in x and 'Se' not in x and 'Sg' not in x and 'Sm' not in x)
plt.figure(figsize=(10, 10))
# 
plt.subplot(2, 2, 1)
plt.scatter(df[~df['Contains_B_O']]['Band Gap (eV)'], df[~df['Contains_B_O']]['Debye T (K)'], marker='o', edgecolor='black', linewidth=0.5, facecolors='gray', alpha=0.5, s=40)
plt.scatter(df[df['Contains_B_O']]['Band Gap (eV)'], df[df['Contains_B_O']]['Debye T (K)'], label='Borates', marker='o', edgecolor='black', linewidth=0.5, color='cornflowerblue', s=40)
plt.scatter(df[df['Contains_S']]['Band Gap (eV)'], df[df['Contains_S']]['Debye T (K)'], label='Sulfides', marker='o', edgecolor='black', linewidth=0.5, color='tab:purple', s=40)
#plt.title('Band Gap vs Debye T')
#plt.xlabel('Band Gap (eV)')
plt.ylabel('Debye T (K)')
plt.ylim(0,1250)
plt.xlim(0,9)
plt.xticks(range(0, 10, 2))  #  X  1  9 2
plt.yticks(range(0,1250, 400))  #  X  1  9 2
plt.legend()
#plt.show()


import re
df['Contains_N'] = df['Formula'].apply(lambda x: bool(re.search(r'N[^a-z]', x)))
def contains_BO_and_halogen_no_lowercase(formula):
    return bool(re.search(r'(Cl|Br|I)(?![a-z])', formula))

df['Contains_O_Cl'] = df['Formula'].apply(contains_BO_and_halogen_no_lowercase)
# 
plt.subplot(2, 2, 2)
plt.scatter(df[~df['Contains_N']]['Band Gap (eV)'], df[~df['Contains_N']]['Debye T (K)'], marker='o', edgecolor='black', linewidth=0.2, facecolors='gray', alpha=0.5, s=40)
plt.scatter(df[df['Contains_N']]['Band Gap (eV)'], df[df['Contains_N']]['Debye T (K)'], label='Nitrides', marker='o', edgecolor='black', linewidth=0.5, color='tab:red', s=40)
plt.scatter(df[df['Contains_O_Cl']]['Band Gap (eV)'], df[df['Contains_O_Cl']]['Debye T (K)'], label='Oxy-halides', marker='o', edgecolor='black', linewidth=0.5, color='wheat', s=40)
#plt.title('Band Gap vs Debye T')
#plt.xlabel('Band Gap (eV)')
plt.ylabel('Debye T (K)')
plt.ylim(0,1250)
plt.xlim(0,9)
plt.xticks(range(0, 10, 2))  #  X  1  9 2
plt.yticks(range(0,1250, 400))  #  X  1  9 2
plt.legend()


import re
# 
df['Contains_Si'] = df['Formula'].apply(lambda x: 'Si' in x and 'O' in x and 'Os' not in x and 'Og' not in x)
df['Contains_F'] = df['Formula'].apply(lambda x: bool(re.search(r'F[^a-z]', x)))
# 
plt.subplot(2, 2, 3)
plt.scatter(df[~df['Contains_Si']]['Band Gap (eV)'], df[~df['Contains_Si']]['Debye T (K)'], marker='o', edgecolor='black', linewidth=0.5, facecolors='gray', alpha=0.5, s=40)
plt.scatter(df[df['Contains_Si']]['Band Gap (eV)'], df[df['Contains_Si']]['Debye T (K)'], label='Silicates', marker='o', edgecolor='black', linewidth=0.5, color='tab:orange', s=40)
plt.scatter(df[df['Contains_F']]['Band Gap (eV)'], df[df['Contains_F']]['Debye T (K)'], label='Fluorides', marker='o', edgecolor='black', linewidth=0.5, color='limegreen', s=40)
#plt.title('Band Gap vs Debye T')
plt.xlabel('Band Gap (eV)')
plt.ylabel('Debye T (K)')
plt.ylim(0,1250)
plt.xlim(0,9)
plt.xticks(range(0, 10, 2))  #  X  1  9 2
plt.yticks(range(0,1250, 400))  #  X  1  9 2
plt.legend()


import re
# 
df['Contains_Al'] = df['Formula'].apply(lambda x: 'Al' in x and 'O' in x and 'Os' not in x and 'Og' not in x)
df['Contains_P'] = df['Formula'].apply(lambda x: 'P' in x and 'O' in x and 'Os' not in x and 'Og' not in x and 'Pt' not in x and 'Pd' not in x)
# 
plt.subplot(2, 2, 4)
plt.scatter(df[~df['Contains_Al']]['Band Gap (eV)'], df[~df['Contains_Al']]['Debye T (K)'], marker='o', edgecolor='black', linewidth=0.5, facecolors='gray', alpha=0.5, s=40)
plt.scatter(df[df['Contains_Al']]['Band Gap (eV)'], df[df['Contains_Al']]['Debye T (K)'], label='Aluminates', marker='o', edgecolor='black', linewidth=0.5, color='cyan', s=40)
plt.scatter(df[df['Contains_P']]['Band Gap (eV)'], df[df['Contains_P']]['Debye T (K)'], label='Phosphates', marker='o', edgecolor='black', linewidth=0.5, color='g', s=40)
#plt.title('Band Gap vs Debye T')
plt.xlabel('Band Gap (eV)')
plt.ylabel('Debye T (K)')
plt.ylim(0,1250)
plt.xlim(0,9)
plt.xticks(range(0, 10, 2))  #  X  1  9 2
plt.yticks(range(0,1250, 400))  #  X  1  9 2
plt.legend()


plt.tight_layout()
plt.show()
#plt.savefig("1.png",dpi=300)
```
